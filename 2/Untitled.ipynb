```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
image = cv2.imread('tv.jpeg' , cv2.IMREAD_GRAYSCALE)
rows , columns = image.shape
print(image.shape)
plt.imshow(image,cmap = 'gray')
```

```python
roi = image[rows //2 -30 : rows // 2 + 160 , columns //2 -170 : columns //2 + 100]
roi_rows , roi_columns = roi.shape
plt.imshow(roi , cmap = 'gray')
```

```python
number = 1

for i in range(1,roi_rows-1):
    for j in range(1,roi_columns-1):
        rndm = random.randint(0,5)

        if rndm < number :
            roi[i,j] = 0
        elif rndm > number:
            roi[i,j] = 255

plt.imshow(roi,cmap = 'gray')
```

```python
image[rows //2 -30 : rows // 2 + 160 , columns //2 -170 : columns //2 + 100] = roi

plt.imshow(image,cmap = 'gray')
cv2.imwrite('result.jpg' , image)
```
